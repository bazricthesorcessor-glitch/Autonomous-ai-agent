```python
import torch
import transformers

print(torch.__version__)
print(transformers.__version__)
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="/kaggle/input/models/google/gemma-4/transformers/gemma-4-26b-a4b-it/1",
    device_map="auto"
)

print("Gemma 4 26B loaded")
```

```python
import gradio as gr

def chat(message):
    result = pipe(
        message,
        max_new_tokens=50
    )[0]["generated_text"]

    return result

demo = gr.Interface(
    fn=chat,
    inputs="text",
    outputs="text"
)

demo.launch(share=True)
```
